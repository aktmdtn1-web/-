```python
# 문자열은 어떻게 만들고 사용할까?

## 1. 큰따옴표로 양쪽 둘러싸기
a = "Hello World"

## 2. 작은 따옴표로 양쪽 둘러싸기
b = 'Python is fun'

## 3. 큰따옴표 3개로 둘러싸기 
c = """Life is too short, You need python"""

## 4. 작은따옴표 3개로 둘러싸기
d = '''Life is too short, You need python'''

print(a)
print(b)
print(c)
print(d) 




```

```python
# 문자열 안에 작은 따옴표를 포함하기
a = "Python's favorite food is perl"
b = """Python's favorite food is perl"""
c = 'Python\'s favorite food is perl'

print(c)
```

```python
# 문자열 안에 큰 따옴표 포함히기
a = 'Python is very easy" he says.'
a = "\"Python is very easy\"he says."

print(a)
```

```python
multiline = """Life is too short
you need Python."""
multiline2 = "Life is too short, \nyou need Python"

print(multiline)
print(multiline2)
```

```python
# 이스케이프 코드
\n -> 문자열 안에서 줄을 바꿀경우
\t -> 문자열 안에서 간격을 둘 때
\\ -> 문자열 안에서 역슬래시를 표현할 경우
\' -> 문자열 안에서 작은 따옴표를 표현할 때
\" -> 문자열 안에서 큰따옴표를 표현할 때
```
