```python
# 교집합 구하기
s1 = set([1, 2, 3, 4, 5, 6])
s2 = set([4, 5, 6, 7, 8, 9])
print(s1)
print(s2)

s1.intersection()
```

```python
# 합집합 구하기
print(s1.union(s2))
```

```python
# 특정 값 제거하기 - remove
s1 = {1, 2, 3}
s1.remove(2)
s1
```
