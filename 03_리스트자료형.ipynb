```python
# 리스트는 어떻게 만들고 사용할까?

# 리스트명 = [요소1, 요소2, ...]

a = []
b = [1,2,3]
c = ["Life", "is", "too", "short"]
d = [1, 2, "life", "is"]
e = [1, 2, ["Life", "is"]]

print(type(a))
print(type(b))
print(type(c))
print(type(d))
print(type(e))
```

```python
# 리스트와 인덱싱과 슬라이싱
a = [1, 2, 3]
print(a[0])
```

```python
#리스트 정렬 - sort
a = [1, 2, 3, 4]
a.sort()
print(a)
a.sort(reverse=True)
print(a)
```

```python
# 리스트 뒤집기 - reverse
a = ['a', 'b', 'c']
a.reverse()
print(a)
```

```python
# 인덱스 반환 - index
a = ['a', 'b', 'c']
a.index("c")
```

```python
# 리스트에 요소 삽입 - insert
a = [1, 2, 3]
a.insert()
```

```python
# 리스트 요소 제거 - remove
a = [1, 2, 3, 1, 2, 3]

a.remove(3)
a.remove(3)
print(a)
```
